```python
import pandas as pd
df = pd.read_csv(r"C:\Users\baiso\OneDrive\Desktop\aqi india data\city_day.csv"  )
print(df.head())
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
print(df.count())
```

```python
pollutant_cols=['PM2.5','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene','Xylene']
for col in pollutant_cols:
    df[col]=df[col].fillna(df[col].median())
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=['AQI'])
```

```python
avg_aqi = df.groupby("City")["AQI"].mean().sort_values(ascending=False)
avg_aqi
```

```python
print('most polluted city')
print(avg_aqi.idxmax(),avg_aqi.max())
```

```python
print('clenest city')
print(avg_aqi.min())
```
